```python
import pandas as pd
import numpy as np
from spicy import stats
```

```python
file = pd.read_csv(r"E:\DATOS\Downloads\fisica\alverja.csv")
#
file = file.rename(columns={'Medidas ': 'Diametro'})

#Mean
promedio = file['Diametro'].mean()
promedio_redondeado = round(promedio, 3)
print("El promedio es : ", promedio_redondeado)

#Desviación estandar
desviacion_estandar = np.std(file['Diametro'])
desv_red = round(desviacion_estandar,3)
print('Desviación estandar :', desv_red)


```

# Histograma

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")

# Crear un histograma de la columna 'Diametro' con mejoras estéticas
plt.figure(figsize=(10, 6))
sns.histplot(file['Diametro'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Histograma de Diametro de Arverjas', fontsize=16)
plt.xlabel('Diametro', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.axvline(promedio, color='red', linestyle='dashed', linewidth=1)
plt.text(promedio + 0.01, plt.ylim()[1] * 0.9, f'Media: {promedio_redondeado}', color='red')
plt.axvline(promedio + desviacion_estandar, color='green', linestyle='dashed', linewidth=1)
plt.axvline(promedio - desviacion_estandar, color='green', linestyle='dashed', linewidth=1)
plt.text(promedio + desviacion_estandar + 0.01, plt.ylim()[1] * 0.8, f'+1σ: {desv_red}', color='green')
plt.text(promedio - desviacion_estandar - 0.1, plt.ylim()[1] * 0.8, f'-1σ: {desv_red}', color='green')
plt.grid(True)
plt.show()
```
