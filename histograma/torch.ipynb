```python
import pandas as pd
import numpy as np
```

```python
frame = pd.read_csv(r"C:\Users\Estudiante\Downloads\Taller\DiabetesK.csv")
print(frame.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(frame['Glucose'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Histograma de glucosa', fontsize=16)
plt.xlabel('Diametro', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
#plt.axvline(promedio, color='red', linestyle='dashed', linewidth=1)
#plt.text(promedio + 0.01, plt.ylim()[1] * 0.9, f'Media: {promedio_redondeado}', color='red')
#plt.axvline(promedio + desviacion_estandar, color='green', linestyle='dashed', linewidth=1)
#plt.axvline(promedio - desviacion_estandar, color='green', linestyle='dashed', linewidth=1)
#plt.text(promedio + desviacion_estandar + 0.01, plt.ylim()[1] * 0.8, f'+1σ: {desv_red}', color='green')
#plt.text(promedio - desviacion_estandar - 0.1, plt.ylim()[1] * 0.8, f'-1σ: {desv_red}', color='green')
plt.grid(True)
plt.show()
```

```python
mediana = round(frame['Glucose'].median(),3)
media = round(frame['Glucose'].mean(),3)
dsv = round(np.std(frame['Glucose']),3)

print("La mediana es {}\nLa media es {}\nLa desviación estandar es{}".format(mediana,media,dsv))
```
